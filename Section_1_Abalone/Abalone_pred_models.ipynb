**Preprocessing**

```python
import pandas as pd
import matplotlib.pyplot as plt
# define the column names 
features = ['sex', 'length', 'diameter', 
            'height','whole weight',
            'shucked weight', 'viscera weight',
           'shell weight', 'rings']

# create a dataframe
abalone_df = pd.read_csv('./Data/abalone.data', sep=',', names = features) 
```

```python
# first 5 from the dataframe
abalone_df.head()
```

```python
# check if any of the column values are missing. 
abalone_df.isnull().sum()
```

```python
# check the dtypes, identify visually if something is continuous, e.g., "sex": 
abalone_df.dtypes
```

```python
from sklearn import preprocessing
```

```python
# use preproccing labelencoder to map 'sex' into R-space. 
le = preprocessing.LabelEncoder() 
le.fit(abalone_df['sex'].unique())

# transform the abalone 'sex' info 
abalone_df['sex'] = le.transform(abalone_df['sex'])
```

```python
# display the abalone_df to show that the "sex" column is mapped to R-space. 
abalone_df.head()
```

```python
# ensure that sex should now be mapped to integer values 
abalone_df.dtypes['sex']
```

```python
# create the age column, where the age is the number of rings * 1.5 
abalone_df['age'] = abalone_df['rings'] + 1.5
```

**Exploration** 

```python
# This section was purely done for fun. 
# I used the features to create a scatter plot to identify the relation between different features. 
# This was limited to 2D space. We see there are features such as 'age' and 'diameter' that are positively correlated. 
# where age increase, so does diameter. 
colors = pd.factorize(abalone_df['age'])[0]
fig = pd.plotting.scatter_matrix(abalone_df, figsize=(20,20), c=colors, s=100,);
plt.savefig('./abalone_feature_plot.png');
```

```python
# boxplot to show the distribution of the numerical physical features, exluding 'sex', 'rings', 'age'. 
abalone_df[features[1:len(features)-1]].boxplot(figsize=(10,10))
plt.savefig('./abalone_box_plot.png');
```

**Regression**  

**NOTE:**  
I am using the convention that X tends to represent the entirety of a feature set. and y represents labels. X and y 
are later split into X_train, X_test, y_train, and y_test by conventional naming standards. 

```python
# seperate the info into X-feature set, and y-labels.
X = abalone_df[features]
y = abalone_df['age']

# Notice from the boxplot above, the whole weight is the sum of 'shucked weight' + 'viscera weight' + 'shell weight'
# therefore, I will remove whole weight as it is not an independent feature. 
X = X.drop(columns='whole weight') 
```

```python
from sklearn.model_selection import train_test_split
```

```python
# split X and y into training and testing sets. 
X_train, X_test, y_train, y_test  = train_test_split(X, y, test_size=.4)
```

```python
from sklearn.tree import DecisionTreeRegressor
# build a Decision tree classifier (dtg_clf). 
dtg_clf = DecisionTreeRegressor(criterion='friedman_mse', 
                           max_depth=3)

# fite the model 
dtg_clf.fit(X_train, y_train) 

# get the predictions 
pred = dtg_clf.predict(X_test) 
```

```python
# having removed the 'whole weight' column, the score went up'
dtg_clf.score(X_test, y_test)
```

```python
# write the instance, prediction value, and actual label to a file. 
f = open('DecTreeRegr_result', 'w') 
f.write('test_instance, regression_pred, ground truth\n') 
for i in range(len(pred)): 
    f.write('{}, {}, {}\n'.format(y_test.index[i], round(pred[i], 3), y_test.to_list()[i]))
f.close() 
```

**Optimize the regression model**   
Decision trees are renowned for overfitting; therefore, I will apply cross-validation for performing GridSearch. GridSearch will do an exhaustive search on the estimator (in this case DecisionTreeRegressor). the parameters are optimized via cross validated gridserach over the parameter grid (I use 'min_samples_split'). For the scorer, I use an r2-scorer, which is a classical means of scoring regression functions. 

**RESOURCES USED:**  
* https://www.programcreek.com/python/example/75177/sklearn.cross_validation.cross_val_score  
* https://stackoverflow.com/questions/44682307/using-decision-tree-regression-and-cross-validation-in-sklearn

```python
from sklearn.model_selection import GridSearchCV 
from sklearn.metrics import make_scorer 
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
```

```python
gridsearch_cv = GridSearchCV(DecisionTreeRegressor(random_state=0),
                            param_grid={'min_samples_split': range(2, 10)}, 
                            scoring=make_scorer(r2_score), cv=3, refit=True) 
# fit the DecisionTree using gridsearch
gridsearch_cv.fit(X_train, y_train) 

# show the r2 score of the best estimator being used on the test set 
r2_score(y_test, gridsearch_cv.best_estimator_.predict(X_test))
```

**Multiclass**  
Logistic Regression is thought of as a probabilistic classification model. It can work effectively on labeling data with a countable number of labels. In this case, we have a range of ages, where 1 < age < 31. Since age can actually be a decimal value, I first converted the age column via a label encoder. Thus mapping the labels to a non-continuous space. 

NOTE:
I am using the convention that X tends to represent the entirety of a feature set. and y represents labels. X and y are later split into X_train, X_test, y_train, and y_test by conventional naming standards.

```python
# use preproccing labelencoder to map 'age' into R-space.
y_le = preprocessing.LabelEncoder() 
y_le.fit(abalone_df['age'].unique())

# transform the abalone 'age' info 
abalone_df['age'] = y_le.transform(abalone_df['age'])
```

```python
# seperate the info into X-feature set, and y-labels.
X = abalone_df[features]
y = abalone_df['age']

# unlike the regression model, this time I do not drop the weight column in X, this improved the model performance. 
```

```python
# split the data ino training and testing. 
X_train, X_test, y_train, y_test  = train_test_split(X, y, test_size=.3)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
import warnings 
warnings.filterwarnings('ignore') 

# build a logistic regression, treat as probabilitistic multinomial classifer. 
log_reg_clf = LogisticRegression(multi_class='multinomial', solver='newton-cg') 

# fit the model 
log_reg_clf.fit(X_train, y_train) 

# get the predictions. 
pred = log_reg_clf.predict(X_test) 
```

```python
# get the score of the model on the test data. 
log_reg_clf.score(X_test, y_test)
```

```python
# write the instance value, the model prediction, and the actual label to a file. 
f = open('multiclass_logreg_result', 'w') 
f.write('test_instance, logistic_regresion_pred, ground truth\n') 
for i in range(len(pred)): 
    f.write('{}, {}, {}\n'.format(y_test.index[i], round(pred[i], 3), y_test.to_list()[i]))
f.close() 
print('completed writing the output to the file multiclass_logreg_result.') 
```
